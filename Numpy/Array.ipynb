## 一、Numpy基础知识
### 1、Numpy元素
* Numpy的数组不同于Python列表，它必须是相同类型的元素。在类型不匹配时会向上转换(如果可行)，如整型转换为浮点型

```python
import numpy as np
x=np.array([1,2.0,3])
type(x[0])
```

* 还可以为数组指定数据类型，采用dtype关键字

```python
import numpy as np
x=np.array([1.0,2,3],dtype='float32')
type(x[1])
np.min(x)
```

#### 创建数组的方法 
* np.ones((n,m),dtype),np.zeros((n,m),dtype):$n\times m$的全1或全0矩阵
* np.full((n,m),x):值全为x的$n\times m$矩阵
* np.linspace(start,end,num):在start与end之间创建num个等距的数
* np.eye(n):n阶单位矩阵
* np.random.randint(st,en,(n,m)):在\[st,en\]之间随机生成$n\times m$矩阵

```python
import numpy as np
one=np.ones((3,3))
eye=np.eye(3)
ran=np.random.randint(0,10,(3,3))
print(one)
print(eye)
print(ran)
```
